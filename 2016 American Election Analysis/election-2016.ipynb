* In order to take only the names for each column (CMTE_ID etc.), we will use BeautifulSoup and requests packages.


```python
r = requests.get("https://www.fec.gov/campaign-finance-data/pac-and-party-summary-file-description/")
pac_soup = BeautifulSoup(r.content)
print(pac_soup)
```

* After inspecting the element "CMTE_ID", we saw that it is the first **td** tag on every **tr** tag(excepting the first row).<br>
* So, we will take the second td element for every table row and we will add it to a list.

```python
tbody = pac_soup.find('table', attrs={'class': 'simple-table dense-table'})
print(tbody)
```

```python
pac_header = []
for row in tbody.find_all("tr"):
    col = row.find("td").contents
    pac_header.extend(col)
pac_header.pop(0) #delete the first row because we don't need it.
pac_header[10] = 'CAND_LOANS' #change from ' CAND_LOANS' to 'CAND_LOANS'.
print(pac_header)
```

```python
pac_df = pd.read_csv('https://www.fec.gov/files/bulk-downloads/2016/webk16.zip',
                     sep='|',
                     index_col=False,
                     names=pac_header) 
pac_df.head(10)
```

* Now we can get the PACs that are supporting Hillary Clinton.

```python
hillary_s = hil_trump.loc[(hil_trump['CAND_NAME'] == 'CLINTON, HILLARY RODHAM') & (hil_trump['SUP_OPP'] == 'S')]
hillary_s.reset_index(drop=True)
trump_o = hil_trump.loc[(hil_trump['CAND_NAME'] == 'TRUMP, DONALD J.') & (hil_trump['SUP_OPP'] == 'O') &
                        (hil_trump['ELE_TYPE'] != 'P')]
trump_o.reset_index(drop=True)
```

* We also combine the new DataFrames, because later we will need some columns.

```python
hillary_full = hillary_s.append(trump_o)
hillary_full.reset_index(drop=True)
```

* Then, we will get the total expenditures for Hillary's supporters and Trump's opposers accordingly.
* We will also drop the duplicates, because othewise we will get a DataFrame for each expenditure with a new column of the total expenditures.
* We need only the CMTE_ID and the total expenditures.

```python
hillary_s_sum = hillary_s.groupby('CMTE_ID').EXP_AMO.agg(['sum']).reindex(hillary_s.CMTE_ID).reset_index()
trump_o_sum = trump_o.groupby('CMTE_ID').EXP_AMO.agg(['sum']).reindex(trump_o.CMTE_ID).reset_index()
hillary_s_sum = hillary_s_sum.drop_duplicates()
trump_o_sum = trump_o_sum.drop_duplicates()
```

* Then we merge the hillary_s_sum and the trump_o_sum, so we can have them on one DataFrame.
* And we fill NaN rows with 0, because some PACs may not make any expenditure for one of the two candidates.

```python
hillary_s = pd.merge(hillary_s_sum,
                      trump_o_sum,
                      on='CMTE_ID',
                     how='outer'
                     )
hillary_s = hillary_s.fillna(0)
```

```python
hillary_s
```

* We need more information about the PACs, like their Type, Total Receipts and Total Disburments.
* So, we will merge the hillary_full DataFrame, which has the Trump's opposers and Hillary's supporters, with pac_df DataFrame, which has more information about the PACs.

```python
pac_hillary = pd.merge(hillary_full[['CMTE_ID','CMTE_NM','CAND_ID','CAND_NAME','SUP_OPP']], 
                       pac_df[['CMTE_ID','CMTE_TP','TTL_RECEIPTS','TTL_DISB']],
                       on='CMTE_ID', 
                       how='inner')
pac_hillary = pac_hillary.drop_duplicates(subset='CMTE_ID', keep="first")
```

* Finally we merge the hillary_s and pac_hillary.
* We are doing a right merge because we need the DataFrame is based on hillary_s (sum's DataFrame) CMTE_ID

```python
hillary = pd.merge(pac_hillary, hillary_s, how='right',on='CMTE_ID' )
hillary
```

* We make a new column on our new DataFrame, the 'TTL_EXP', so we can measure how much did every PAC spent.
* We add the sum_x (Hillary's supporters) and sum_y (Trump's opposers)

```python
hillary['TTL_CAMP'] = hillary['sum_x'] + hillary['sum_y']
```

* We found out that the dates on our data set are not correct.
* We will make the changes only the dataframes that we made.
* First, we will work on the hil_exp and tru_exp.
* We need all of our dates have lenght = 10. So, we will take only those rows.

```python
hil_exp = hil_exp.loc[ (hil_exp['TRANSACTION_DT'].str.len() == 10)]
tru_exp = tru_exp.loc[ (tru_exp['TRANSACTION_DT'].str.len() == 10)]
```

* Then we can change the dtype of the TRANSACTION_DT to datetime.

```python
hil_exp['TRANSACTION_DT'] = pd.to_datetime(hil_exp['TRANSACTION_DT'])
tru_exp['TRANSACTION_DT'] = pd.to_datetime(tru_exp['TRANSACTION_DT'])
```

* Also, we will delete any other year than 2015 and 2016

```python
hil_exp = hil_exp.loc[ (hil_exp['TRANSACTION_DT'].dt.year == 2015) |
                                         (hil_exp['TRANSACTION_DT'].dt.year == 2016) ]
tru_exp = tru_exp.loc[ (tru_exp['TRANSACTION_DT'].dt.year == 2015) |
                                         (tru_exp['TRANSACTION_DT'].dt.year == 2016) ]
```

* We found out that individual expeditures dataframe has wrong dates.
* So, we will do the same for hil_ra and tru_ra dataframes.
* The only diffence is that days, months and years are not seperated with / .
* So we need the transactions dates that have length = 8.
* Also, because there are some values that are not in 20XX format, we will choose only the values that their 4 last digits are 2016 or 2015

```python
y_value = ['2015','2016']
```

```python
hil_ra = hil_ra.loc[ (hil_ra['TRANSACTION_DT'].str.len() == 8) &
                   (hil_ra['TRANSACTION_DT'].str.slice(start=4, stop=8).isin(y_value))]
tru_ra = tru_ra.loc[ (tru_ra['TRANSACTION_DT'].str.len() == 8) &
                   (tru_ra['TRANSACTION_DT'].str.slice(start=4, stop=8).isin(y_value))]
```

* Now we can change the data type of our dates.

```python
hil_ra['TRANSACTION_DT'] = pd.to_datetime(hil_ra['TRANSACTION_DT'], format='%m%d%Y')
tru_ra['TRANSACTION_DT'] = pd.to_datetime(tru_ra['TRANSACTION_DT'], format='%m%d%Y')
```

* Now that we have correctly set our data types, we can investigate the chronological evolution of our data.
* We will re-index our data sets, because we want to groupby them by month.

```python
hil_ra.index = hil_ra['TRANSACTION_DT']
hil_exp.index = hil_exp['TRANSACTION_DT']

tru_ra.index = tru_ra['TRANSACTION_DT']
tru_exp.index = tru_exp['TRANSACTION_DT']
```

* Now we will groupby hil_ra by TRANSACTION_DT, and we need the sum of TRANSACTION_AMT for every group.
* We do the same for the dataframe hil_exp.

```python
hil_ra_group = hil_ra.groupby(pd.Grouper(key='TRANSACTION_DT',freq="M")).sum('TRANSACTION_AMT')
hil_exp_group = hil_exp.groupby(pd.Grouper(key='TRANSACTION_DT',freq="M")).sum('TRANSACTION_AMT')
```

* Now we will merge the new dataframes.
* We will merge on TRANSACTION_DT, and will use outer, because a month can have no TRANSACTION_AMT.

```python
hil_total_mr = pd.merge(hil_ra_group,
                        hil_exp_group,
                        on='TRANSACTION_DT',
                        how='outer')
hil_total_mr
```
